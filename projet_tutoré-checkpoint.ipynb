```python
import pandas as pd
import matplotlib.pyplot as plt

# Lire et préparer les données
files = ['beijing.csv', 'marseille.csv', 'berlin.csv', 'madrid.csv', 'london.csv', 'lisbonne.csv', 'paris.csv', 'rome.csv', 
         'bangkok .csv', 'jerusalem.csv', 'monterrey.csv', 'ottawa.csv', 'new_york.csv', 'new_delhi.csv', 'teheran.csv',
         'washington.csv', 'sao_paulo.csv', 'tokyo.csv', 'abidjan.csv', 'abuja.csv', 'bamako.csv', 'lome.csv', 'n_djamena.csv']

cities_info = {
    'beijing.csv': ('Beijing', 'Asie'),
    'marseille.csv': ('Marseille', 'Europe'),
    'berlin.csv': ('Berlin', 'Europe'),
    'madrid.csv': ('Madrid', 'Europe'),
    'london.csv': ('London', 'Europe'),
    'lisbonne.csv': ('Lisbonne', 'Europe'),
    'paris.csv': ('Paris', 'Europe'),
    'rome.csv': ('Rome', 'Europe'),
    'bangkok .csv': ('Bangkok', 'Asie'),
    'jerusalem.csv': ('Jerusalem', 'Asie'),
    'monterrey.csv': ('Monterrey', 'Amérique du Sud'),
    'ottawa.csv': ('Ottawa', 'Amérique du Nord'),
    'new_york.csv': ('New_york', 'Amérique du Nord'),
    'new_delhi.csv': ('New_delhi', 'Asie'),
    'teheran.csv': ('Teheran', 'Asie'),
    'washington.csv': ('Washington', 'Amérique du Nord'),
    'sao_paulo.csv': ('Sao_Paulo', 'Amérique du Sud'),
    'tokyo.csv': ('Tokyo', 'Asie'),
    'abidjan.csv': ('Abidjan', 'Afrique'),
    'abuja.csv': ('Abuja', 'Afrique'),
    'bamako.csv': ('Bamako', 'Afrique'),
    'lome.csv': ('Lomé', 'Afrique'),
    'n_djamena.csv': ('N\'Djamena', 'Afrique')
}

dataframes = []
for file in files:
    df = pd.read_csv(file, na_values=['', ' ', 'NA', 'N/A', 'NULL'])
    ville, continent = cities_info[file]
    df['Ville'] = ville
    df['Continent'] = continent
    dataframes.append(df)

# Combiner tous les DataFrames en un seul
data = pd.concat(dataframes, ignore_index=True)
```

Transformation des valeurs manqauntes en les remplacant par la moyenne des polluants par villes

```python
import pandas as pd
import numpy as np

# Assurez-vous que les noms des colonnes n'ont pas d'espaces au début
data.columns = data.columns.str.strip()

# Liste des polluants
polluants = ['pm25', 'pm10', 'o3', 'no2', 'so2', 'co']

# Fonction pour remplacer les valeurs manquantes par la moyenne de chaque groupe
def remplacer_par_moyenne(groupe):
    return groupe.fillna(groupe.mean())

# Appliquer la fonction à chaque groupe (chaque ville)
data_remplace = data.groupby('Ville')[polluants].transform(remplacer_par_moyenne)

# Mettre à jour le DataFrame original avec les valeurs imputées
data.update(data_remplace)

# Afficher les premières lignes des données avec les valeurs manquantes remplacées par la moyenne
data
```

Determination des seuils critiques et calcul de l'AQI

```python
import pandas as pd

# Définir les seuils AQI et les valeurs AQI correspondantes
seuils_aqi = [
    (0, 11, 50), (12.1, 35, 100), (35.1, 55, 150),
    (55.1, 150, 200), (150.1, 250.4, 300), (250.5, 350.4, 400),
    (350.5, 500.4, 500)
]

def calculer_aqi(nom_polluant, concentration):
    for bas, haut, aqi in seuils_aqi:
        if bas <= concentration <= haut:
            return aqi
    return None

def calculer_aqi_global(ligne):
    valeurs_aqi = []
    polluants = ['co', 'no2', 'o3', 'so2', 'pm25', 'pm10']
    for polluant in polluants:
        aqi = calculer_aqi(polluant, ligne[polluant])
        if aqi is not None:
            valeurs_aqi.append(aqi)
    return max(valeurs_aqi) if valeurs_aqi else None

# Calculer l'AQI pour chaque ligne
data['AQI'] = data.apply(calculer_aqi_global, axis=1)

# Définir les catégories AQI
categories_aqi = [
    (0, 50, 'Bon'), (51, 150, 'Modéré'),
    (151, 200, 'Nocif'), (201, 500, 'Très nocif')
]

def categoriser_aqi(valeur_aqi):
    for bas, haut, categorie in categories_aqi:
        if bas <= valeur_aqi <= haut:
            return categorie
    return None

# Catégoriser l'AQI
data['Catégorie AQI'] = data['AQI'].apply(categoriser_aqi)

# Afficher les premières lignes du DataFrame
data
```

clustering des villes 

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Calculate the mean AQI for each city
city_aqi_mean = data.groupby('Ville')['AQI'].mean().reset_index()

inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(city_aqi_mean[['AQI']])
    inertia.append(kmeans.inertia_)

# Tracer le graphique de la méthode du coude
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, 'bo-')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Inertie')
plt.title('Méthode du coude pour déterminer k optimal')
plt.grid(True)
plt.show()
```

Le graphique de la méthode du coude montre un changement de pente marqué autour de 𝑘=3 .
 Pour notre travil, nous allons choisir 𝑘=3 pour effectuer notre clustering.

Appliquons maintenant le clustering k-means avec 𝑘=3 et visualisons les résultats.

```python
# Apply k-means clustering with k = 3
kmeans = KMeans(n_clusters=3, random_state=42)
city_aqi_mean['Cluster'] = kmeans.fit_predict(city_aqi_mean[['AQI']])

# Display the cluster centers
cluster_centers = kmeans.cluster_centers_
cluster_centers

# Visualize the clustering results
plt.figure(figsize=(10, 6))
for cluster in range(3):
    plt.scatter(
        city_aqi_mean[city_aqi_mean['Cluster'] == cluster]['Ville'],
        city_aqi_mean[city_aqi_mean['Cluster'] == cluster]['AQI'],
        label=f'Cluster {cluster}'
    )

plt.xlabel('City')
plt.ylabel('Average AQI')
plt.title('Clustering of Cities by Average AQI')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()

```

On peut constater voir aisement les villes de chaque cluster, le premeir groupe en bleu represente les villes ayant les plus bonnes qualités d'air, le second groupe 
en vert represente les villes dont l'air est en moyenne nocifs, et le troisieme en orange les villes dont l'air est tres nocif

```python
# Appliquer le clustering hiérarchique
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(city_aqi_mean[['AQI']], method='ward')

# Tracer le dendrogramme
plt.figure(figsize=(12, 8))
dendrogram(Z, labels=city_aqi_mean['Ville'].values, leaf_rotation=90, leaf_font_size=10)
plt.title('Dendrogramme des villes basé sur l\'AQI moyen')
plt.xlabel('Ville')
plt.ylabel('Distance')
plt.show()
```

Le dendogramme nous permet de visualiser les villes ayant les qualités d'air similaires 

Vérification des conditions pour l'application du modèle ARIMA

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# Charger les données
data = pd.read_csv('air_q.csv', parse_dates=['date'], index_col='date')

# Visualisation des données
plt.figure(figsize=(10, 6))
plt.plot(data['AQI'])
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('Évolution de l\'AQI au fil du temps')
plt.show()

# Test de stationnarité
result = adfuller(data['AQI'])
print('Test Augmented Dickey-Fuller (ADF) :')
print(f'Statistique de test ADF : {result[0]}')
print(f'p-valeur : {result[1]}')
print('Résultat du test :')
if result[1] <= 0.05:
    print('Les données sont stationnaires.')
else:
    print('Les données ne sont pas stationnaires.')

# Autocorrélation et autocorrélation partielle
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
plot_acf(data['AQI'], lags=30, ax=axes[0])
plot_pacf(data['AQI'], lags=30, ax=axes[1])
plt.show()

# Séparation des données en ensembles d'entraînement et de test
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size]
test_data = data.iloc[train_size:]

# Ajustement du modèle ARIMA
model = ARIMA(train_data['AQI'], order=(5, 1, 0))
model_fit = model.fit()

# Prédictions
predictions = model_fit.forecast(steps=len(test_data))

# Évaluation du modèle
rmse = ((test_data['AQI'] - predictions) ** 2).mean() ** 0.5
print(f'RMSE : {rmse}')

# Visualisation des prédictions
plt.figure(figsize=(10, 6))
plt.plot(test_data.index, test_data['AQI'], label='Valeurs réelles')
plt.plot(test_data.index, predictions, label='Prédictions ARIMA')
plt.xlabel('date')
plt.ylabel('AQI')
plt.title('Prédictions du modèle ARIMA')
plt.legend()
plt.show()
```
